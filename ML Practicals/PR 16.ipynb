```python
# Attempt following questions based on matplotlib library
# a. Write a program to plot a line graph showing the relationship between two lists of values.
# b. Write a program to draw a line graph and add: Title, X-axis label, Y-axis label
# c. Write a Program to plot two or more lines on the same graph with labels.
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [15, 25, 35, 45, 55]
# a. Plot a simple line graph showing relationship between two lists
plt.plot(x, y1)
plt.show()
# b. Line graph with Title, X-axis label, and Y-axis label
plt.plot(x, y1)
plt.title("Line Graph Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
# c. Plot two lines on the same graph with labels
plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.title("Multiple Line Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.show()
```
