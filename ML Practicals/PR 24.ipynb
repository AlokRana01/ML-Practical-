```python
# Write a Python program to apply Linear Regression using Scikit-learn and visualize the data and regression line using Seaborn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Step 1: Create sample dataset
data = {
"Hours_Studied": [1, 2, 3, 4, 5, 6, 7, 8],
"Marks": [35, 40, 45, 50, 55, 60, 65, 70]
}
df = pd.DataFrame(data)
# Step 2: Prepare data for model
X = df[["Hours_Studied"]] # Independent variable (2D)
y = df["Marks"] # Dependent variable
# Step 3: Apply Linear Regression
model = LinearRegression()
model.fit(X, y)
# Predict values
df["Predicted_Marks"] = model.predict(X)
# Step 4: Visualize using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
# Scatter plot of actual data
sns.scatterplot(x="Hours_Studied", y="Marks", data=df, label="Actual Data")
# Regression line
sns.lineplot(x="Hours_Studied", y="Predicted_Marks", data=df, color="red",
label="Regression Line")
plt.title("Linear Regression: Hours Studied vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
