```python
# Plot line and bar graph together using subplots.
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
line_values = [10, 20, 30, 40, 50]
bar_values = [5, 15, 25, 35, 45]
# Create figure
plt.figure(figsize=(8, 6))
# Line graph (subplot 1)
plt.subplot(2, 1, 1)
plt.plot(x, line_values)
plt.title("Line Graph")
plt.xlabel("X Values")
plt.ylabel("Line Values")
# Bar graph (subplot 2)
plt.subplot(2, 1, 2)
plt.bar(x, bar_values)
plt.title("Bar Graph")
plt.xlabel("X Values")
plt.ylabel("Bar Values")
# Adjust layout and display
plt.tight_layout()
plt.show()
```
