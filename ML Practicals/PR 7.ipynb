```python
# Attempt following questions based on Pandas – DataFrame Basics
# a. Create a DataFrame from a dictionary
# b. Display last 5 rows of a DataFrame
# c. Display column names of a DataFrame
# d. Find total number of rows and columns
import pandas as pd
# a. Create a DataFrame from a dictionary
data = {
"Name": ["Amit", "Neha", "Rahul", "Priya", "Suresh", "Kiran", "Anita"],
"Age": [23, 25, 22, 24, 26, 27, 21],
"City": ["Delhi", "Mumbai", "Pune", "Delhi", "Chennai", "Bangalore", "Jaipur"]
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
print()
# b. Display last 5 rows of the DataFrame
print("Last 5 rows of DataFrame:")
print(df.tail())
print()
# c. Display column names of the DataFrame
print("Column names:")
print(df.columns)
print()
# d. Find total number of rows and columns
rows, columns = df.shape
print("Total number of rows:", rows)
print("Total number of columns:", columns)
```
