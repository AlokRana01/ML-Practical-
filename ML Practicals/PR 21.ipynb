```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y_line = [10, 20, 30, 40, 50]
y_bar = [5, 15, 25, 35, 45]

plt.figure(figsize=(8, 6))

# First graph – Pie chart (instead of line graph)
plt.subplot(1,2, 1)
plt.pie(y_line, labels=x, autopct='%1.1f%%')
plt.title("Pie Chart")

# Second graph – Bar graph
plt.subplot(1,2 , 2)
plt.bar(x, y_bar)
plt.title("Bar Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.tight_layout()
plt.show()
```
