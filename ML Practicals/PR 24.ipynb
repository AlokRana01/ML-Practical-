```python
# Write a Python program to apply Linear Regression using Scikit-learn and visualize the data and regression line using Seaborn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([2,4,5,4,5])

model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

plt.scatter(x,y)
plt.plot(x, y_pred)
plt.title("Linear Regression ")
plt.xlabel("X value ")
plt.ylabel("Y value")
plt.show()
```
