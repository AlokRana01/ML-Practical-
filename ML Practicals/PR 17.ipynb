```python
# Plot a line graph showing monthly sales data of a company.
import matplotlib.pyplot as plt
# Monthly sales data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
"Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [12000, 15000, 18000, 17000, 20000, 22000,
21000, 23000, 25000, 24000, 26000, 28000]
# Plot line graph
plt.plot(months, sales)
plt.title("Monthly Sales Data of a Company")
plt.xlabel("Months")
plt.ylabel("Sales Amount")
plt.show()
```
