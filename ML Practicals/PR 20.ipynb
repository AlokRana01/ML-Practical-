```python
# Attempt following questions based on matplotlib library
# a. Write a program to display a bar chart showing marks obtained by students in different subjects.
# b. Write a program to draw a horizontal bar chart using Matplotlib.
# c. Write a program to create a pie chart representing percentage distribution of data.
import matplotlib.pyplot as plt
# a. Bar chart – Marks in subjects
subjects = ["Maths", "Science", "English", "Computer", "History"]
marks = [85, 78, 90, 88, 75]
plt.bar(subjects, marks)
plt.title("Marks Obtained by Students in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
# b. Horizontal bar chart
plt.barh(subjects, marks)
plt.title("Horizontal Bar Chart of Marks")
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.show()
# c. Pie chart – Percentage distribution
labels = ["Maths", "Science", "English", "Computer", "History"]
sizes = [85, 78, 90, 88, 75]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Percentage Distribution of Marks")
plt.show()
```
