```python
# Plot two different graphs in the same figure.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y_line = [10, 20, 30, 40, 50]
y_bar = [5, 15, 25, 35, 45]
# Create a figure with 2 subplots
plt.figure(figsize=(8, 6))
# First graph – Line graph
plt.subplot(2, 1, 1)
plt.plot(x, y_line)
plt.title("Line Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
# Second graph – Bar graph
plt.subplot(2, 1, 2)
plt.bar(x, y_bar)
plt.title("Bar Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
# Display the figure
plt.tight_layout()
plt.show()
```
