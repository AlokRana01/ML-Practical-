```python
# Draw a line graph to compare temperatures of two cities over a week.
import matplotlib.pyplot as plt
# Data for one week
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
city1_temps = [22, 24, 19, 23, 25, 27, 26]
city2_temps = [18, 20, 17, 21, 22, 24, 23]
# Plot line graph
plt.plot(days, city1_temps, marker='o', label="City 1")
plt.plot(days, city2_temps, marker='o', label="City 2")
# Add title and labels
plt.title("Weekly Temperature Comparison of Two Cities")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.legend()
# Display graph
plt.show()
```
