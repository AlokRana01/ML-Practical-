```python
# Write a Python program to implement the K-Nearest Neighbors (KNN) algorithm with K = 3 and classify a new data point using a simple dataset.
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
# Step 1: Create a simple dataset
# Feature: Marks
X = np.array([[35], [45], [55], [65], [75], [85]])
# Labels: 0 = Fail, 1 = Pass
y = np.array([0, 0, 1, 1, 1, 1])
# Step 2: Create KNN model with K = 3
knn = KNeighborsClassifier(n_neighbors=3)
# Train the model
knn.fit(X, y)
# Step 3: Classify a new data point
new_data = np.array([[50]]) # New student's marks
prediction = knn.predict(new_data)
# Step 4: Display result
if prediction[0] == 1:
  print("New Data Point Result: Pass")
else:
  print("New Data Point Result: Fail")
```
