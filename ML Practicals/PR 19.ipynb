```python
# Plot two lines on the same graph with different colors and styles.
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [50, 40, 30, 20, 10]
# Plot two lines with different colors and styles
plt.plot(x, y1, color='blue', linestyle='-', label='Line 1')
plt.plot(x, y2, color='red', linestyle='--', label='Line 2')
# Add title and labels
plt.title("Two Lines with Different Colors and Styles")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
# Display graph
plt.show()
```
